```python
import pandas as pd
from faker import Faker

fake = Faker()

# Generate customers
customers = []
for i in range(10):
    customers.append({
        'CustomerID': i,
        'Name': fake.name(),
        'Email': fake.email()
    })

# Generate customer addresses
customer_addresses = []
for i in range(20):
    customer_addresses.append({
        'AddressID': i,
        'CustomerID': i % 10,
        'Address': fake.address()
    })

# Generate products
products = []
for i in range(10):
    products.append({
        'ProductID': i,
        'Name': fake.bs(),
        'Description': fake.text()
    })

# Generate product variants
product_variants = []
for i in range(20):
    product_variants.append({
        'VariantID': i,
        'ProductID': i % 10,
        'Attribute1': fake.word(),
        'Attribute2': fake.word()
    })

# Generate product pricing
product_pricing = []
for i in range(30):
    product_pricing.append({
        'PricingID': i,
        'VariantID': i % 20,
        'Price': fake.random_int(10, 100),
        'EffectiveDate': fake.date_between(start_date='-2y', end_date='today')
    })

# Generate orders
orders = []
for i in range(50):
    orders.append({
        'OrderID': i,
        'CustomerID': i % 10,
        'OrderDate': fake.date_between(start_date='-2y', end_date='today')
    })

# Generate order items
order_items = []
for i in range(100):
    order_items.append({
        'ItemID': i,
        'OrderID': i % 50,
        'VariantID': i % 20,
        'Quantity': fake.random_int(1, 5)
    })

# Create dataframes
customers_df = pd.DataFrame(customers)
customer_addresses_df = pd.DataFrame(customer_addresses)
products_df = pd.DataFrame(products)
product_variants_df = pd.DataFrame(product_variants)
product_pricing_df = pd.DataFrame(product_pricing)
orders_df = pd.DataFrame(orders)
order_items_df = pd.DataFrame(order_items)

# Save data to CSV files
customers_df.to_csv('customers.csv', index=False)
customer_addresses_df.to_csv('customer_addresses.csv', index=False)
products_df.to_csv('products.csv', index=False)
product_variants_df.to_csv('product_variants.csv', index=False)
product_pricing_df.to_csv('product_pricing.csv', index=False)
orders_df.to_csv('orders.csv', index=False)
order_items_df.to_csv('order_items.csv', index=False)
```
